```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean 
```

```python
df = pd.read_csv("../dataset_live/expanded/load_pwr.csv")
df.head()
```

```python
df.describe()
```

```python
df["daylight_time"].unique()
```

```python
df_hour = df.groupby("hour", as_index=False).agg("mean")
display(df_hour)
```

```python
df_hour[df_hour["hour"]==23]["meter_0"].values[0]
```

```python
plt.figure(figsize=(15,6))
plt.style.use("dark_background")
sns.lineplot(x="hour", y="meter_0", color="#FF5722", palette="pastel", data=df_hour);
```

```python
df_g = df_hour[["hour"]]
df_g["mean"] = 0
df_g
```

```python
meters = ["meter_%d" % i for i in range(215)]
for h in df_hour.hour:
    values = []
    for m in meters:
        values.append(df_hour[df_hour["hour"]==h][m].values[0])
    df_g.iloc[h,1] = mean(values)
df_g.head()
```

```python
plt.style.use("dark_background")
sns.set_style("darkgrid")
plt.figure(figsize=(15,6))
# sns.set(rc={'axes.facecolor':'#393536', 'figure.facecolor':'#393536'})
sns.lineplot(x="hour", y="mean", color="#fc595c", palette="pastel", data=df_g, sizes=(16,10)).set_title('Daily Average Consumption per hour from all meters', fontsize=16);
```
